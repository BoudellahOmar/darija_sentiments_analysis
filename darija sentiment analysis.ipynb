```python
df.text=df.text.str.lower()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
vect=TfidfVectorizer(max_features=5000)
```

```python
listtweet=df.text.tolist()
```

```python
listtweet[0:5]
```

```python
import re
li=[]
for i in listtweet:
    li.append(re.sub(r'[^A-Za-z]+',' ',str(i)))

li
```

```python
x=vect.fit_transform(li)
dataset=pd.DataFrame(x.toarray(),columns=vect.get_feature_names_out())
dataset
```

```python
dataset.columns
```

```python
ind=[]
col=[]
for i in dataset.columns:
    if len(i)<=2:
        ind.append(1)
    else:
        ind.append(0)
for i in range(len(ind)):
    if ind[i]==1:
        col.append(i)
dataset=dataset.drop(dataset.columns[[col]], axis = 1)

        
```
